```python
import bson
import matplotlib.pyplot as plt
```

```python
report_path = "report.bson"

with open(report_path, "rb") as file:
    report = bson.loads(file.read())
```

```python
W1_weights_change = []

for i in range(len(report["report"])):
    if report["report"][i]["type"] == "backward":
        backward_report = report["report"][i]

        for j in range(len(backward_report["W1_after"]["values"])):
            if j == len(W1_weights_change):
                W1_weights_change.append([])
            
            W1_weights_change[j].append(backward_report["W1_after"]["values"][j])

W2_weights_change = []

for i in range(len(report["report"])):
    if report["report"][i]["type"] == "backward":
        backward_report = report["report"][i]

        for j in range(len(backward_report["W2_after"]["values"])):
            if j == len(W2_weights_change):
                W2_weights_change.append([])
            
            W2_weights_change[j].append(backward_report["W2_after"]["values"][j])
```

```python
fig, ax = plt.subplots()

# for i in range(len(W2_weights_change)):
ax.plot(W2_weights_change[1])

plt.show()
```
